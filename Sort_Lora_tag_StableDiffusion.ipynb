#**Sort Lora tag**

フォルダ内のテキストファイル（カンマ区切りでタグが並べてある）

から、特定のタグを削除する

```python
from google.colab import drive

# Google Driveをマウントする
drive.mount('/content/drive')
```

```python
parent_folder = "/content/drive/MyDrive/Deep_learning/666mai_dataset/Lora-tags-grav"
dst_folder = "/content/drive/MyDrive/Deep_learning/666mai_dataset/Lora-tags-grav-revised"

tags_to_exclude = ["solo", "looking at viewer", "close-up", "realistic", "portrait", "eye-focus", "closed mouth", "male focus"]
```

```python
import os
import shutil

def process_txt_file(file_path):
    with open(file_path, 'r') as file:
        original_tags = file.readline().strip().split(', ')
        processed_tags = [tag for tag in original_tags if tag not in tags_to_exclude]
        processed_text = ', '.join(processed_tags)
        excluded_tags = [tag for tag in original_tags if tag in tags_to_exclude]

    with open(file_path, 'w') as file:
        file.write(processed_text)

    return processed_text, excluded_tags

def process_txt_files_in_folder(folder_path):
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):
            file_path = os.path.join(folder_path, file_name)
            processed_text, excluded_tags = process_txt_file(file_path)
            print(f"Processed file: {file_name}")
            print(f"Processed text: {processed_text}")
            print(f"Excluded tags: {', '.join(excluded_tags)}")
            print("--------")

def main():
    # 出力先フォルダが存在しない場合は作成する
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    # txtファイルを処理して出力先フォルダに保存し、内容と削除されたタグを表示する
    process_txt_files_in_folder(parent_folder)

    print("処理が完了しました。")

main()
```
