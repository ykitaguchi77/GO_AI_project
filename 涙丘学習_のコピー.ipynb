```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# モデルの初期化と読み込み
model_path = '/content/drive/MyDrive/AI_laboratory_course/SBVPI/yolov8/result/custom_train2/weights/best.pt'
model = YOLO(model_path)

# 画像の読み込み
image_path = '/content/drive/MyDrive/AI_laboratory_course/SBVPI/yolov8/dataset/images/valid/13L_s_4.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# セグメンテーションの予測を実行
results = model.predict(image)

# 予測結果の描画
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')

# 検出されたセグメンテーションマスクがあるかどうか確認
if hasattr(results, 'segments') and results.segments.any():
    # セグメンテーションマスクの描画
    alpha = 0.5  # 透明度の設定
    for mask in results.segments:
        plt.imshow(mask, cmap='jet', alpha=alpha)  # セグメンテーションマスクのオーバーレイ

plt.show()
```
