```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# ãƒ¢ãƒ‡ãƒ«ã®åˆæœŸåŒ–ã¨èª­ã¿è¾¼ã¿
model_path = '/content/drive/MyDrive/AI_laboratory_course/SBVPI/yolov8/result/custom_train2/weights/best.pt'
model = YOLO(model_path)

# ç”»åƒã®èª­ã¿è¾¼ã¿
image_path = '/content/drive/MyDrive/AI_laboratory_course/SBVPI/yolov8/dataset/images/valid/13L_s_4.jpg'
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ã‚»ã‚°ãƒ¡ãƒ³ãƒ†ãƒ¼ã‚·ãƒ§ãƒ³ã®äºˆæ¸¬ã‚’å®Ÿè¡Œ
results = model.predict(image)

# äºˆæ¸¬çµæœã®æç”»
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')

# æ¤œå‡ºã•ã‚ŒãŸã‚»ã‚°ãƒ¡ãƒ³ãƒ†ãƒ¼ã‚·ãƒ§ãƒ³ãƒã‚¹ã‚¯ãŒã‚ã‚‹ã‹ã©ã†ã‹ç¢ºèª
if hasattr(results, 'segments') and results.segments.any():
    # ã‚»ã‚°ãƒ¡ãƒ³ãƒ†ãƒ¼ã‚·ãƒ§ãƒ³ãƒã‚¹ã‚¯ã®æç”»
    alpha = 0.5  # é€æ˜åº¦ã®è¨­å®š
    for mask in results.segments:
        plt.imshow(mask, cmap='jet', alpha=alpha)  # ã‚»ã‚°ãƒ¡ãƒ³ãƒ†ãƒ¼ã‚·ãƒ§ãƒ³ãƒã‚¹ã‚¯ã®ã‚ªãƒ¼ãƒãƒ¼ãƒ¬ã‚¤

plt.show()
```
